# import necesary library

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

# my questions with answer

```python
my_faq_data = { "What is AI":"AI means Artificial Intelligence.",
    "What is machine learning":"Machine learning is a subset of AI that learns from data.",
    "What is python":"Python is a programming language.",
    "What is deep learning":"Deep learning is a type of machine learning using neural networks.",
    "What is numpy":"NumPy is a library used for numerical computing.",
    "What is pandas":"Pandas is used for data analysis and manipulation.",
    "What is overfitting":"Overfitting happens when a model learns training data too well.",
    "What is underfitting":"Underfitting happens when a model cannot learn patterns properly.",
    "What is classification":"Classification predicts categories.",
    "What is regression":"Regression predicts continuous values."}


```

```python
my_question = list(my_faq_data.keys())
```

```python
my_question
```

# seperate just the questions

```python
my_question
```

# apply text vectorizer and lower case the text same time to generate same vector for the same data present in lowercase or upper

```python
vec= TfidfVectorizer(encoding='utf-8', lowercase=True)
```

```python
my_question_vec_data= vec.fit_transform(my_question)
```

```python
my_question_vec_data.shape
```

```python
my_question_vec_data.toarray()
```

```python
def get_faq_answer(user_query: str, threshold = .4) -> str:
    query = vec.transform([user_query])
    similarity = cosine_similarity(query, my_question_vec_data).flatten()
    question_match = np.argmax(similarity) # [.8,.75,6]

    score = similarity[question_match]
    print(score) # similarity score will be return

    if score > threshold:
        question = my_question[question_match]
        return my_faq_data[question]
    else:
        return "sorry sir! I don't have any knowledge about it. please ask anything about services"
```

```python
get_faq_answer("What are the facilites you provides")
```
